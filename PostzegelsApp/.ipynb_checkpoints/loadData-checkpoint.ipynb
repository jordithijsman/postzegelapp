```python
import numpy as np
import pandas as pd
import json
import requests as req
```

```python
data = pd.read_csv("PostJohan.csv", sep=";", encoding='cp1252')
#data = data.iloc[0:10] #voor testen
```

```python
first_row = data.iloc[1]
first_row
```

```python
#pre-processing lege cellen
data['Omschrijving'] = data['Omschrijving'].fillna('')
data['Naam'] = data['Naam'].fillna('')
data['Kleur'] = data['Kleur'].fillna('')
data['Tanding'] = data['Tanding'].fillna('')
data['Waarde postfris'] = data['Waarde postfris'].fillna('')
data['Waarde gestempeld'] = data['Waarde gestempeld'].fillna('')
data['Uitgiftedatum'] = data['Uitgiftedatum'].fillna('')
```

```python
def row_to_json(row):
    dataset = { "id": row.Postzegel_id, 
                "omschrijving": row.Omschrijving, 
                "naam": row.Naam, 
                "kleur": row.Kleur, 
                "tanding": row.Tanding, 
                "waarde_postfris": row["Waarde postfris"], 
                "waarde_gestempeld":row["Waarde gestempeld"],
                "land": "BelgiÃ«"
                #,"datum":row.Uitgiftedatum
              }

    #json_dump = json.dumps(dataset)
    #json_obj = json.loads(json_dump)
    return dataset
```

```python
#inlezen van postzegels
for index, row in data.iterrows():
    json_obj = row_to_json(row)
    #print(json_obj)
    response = req.post('http://localhost:3000/postzegel', json = json_obj)
    print("index: %d, status: " %(index), response)
    
    
```
